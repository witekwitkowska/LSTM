```python
with open('/media/usuario/datos/results/' + model, 'rb') as file_pi:
    history = pickle.load(file_pi)
print(history['val_loss'][-1])
print(history['val_acc'][-1])
```

```python

# %matplotlib inline
import pylab as plt
# Plot training & validation accuracy values
fig = plt.figure(figsize=(20, 10))

plt.plot(history['acc'])
plt.plot(history['val_acc'])
plt.title('Model accuracy:' + model)
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper left')
plt.savefig('/media/usuario/datos/results/charts/' + model +  '-acc.png')
plt.show()



# Plot training & validation loss values
fig = plt.figure(figsize=(20, 10))

plt.plot(history['loss'])
plt.plot(history['val_loss'])
plt.title('Model loss' + model)
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper left')
plt.savefig('/media/usuario/datos/results/charts/' + model +  '-loss.png')
plt.show()
```
